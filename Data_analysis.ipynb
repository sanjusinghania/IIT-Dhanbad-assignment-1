```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

csv_file = 'udemy_courses.csv'
df = pd.read_csv(csv_file)
df.head()
```

```python
df = df.drop(['course_id','url','level','content_duration','published_timestamp'], axis=1)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows:" , duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python

df = df.dropna() #no null values present
df.count()
```

```python
df['is_paid'].value_counts()
```

```python
df_notpaid = df[~df["is_paid"]]
df_notpaid
```

```python
df_paid = df[df["is_paid"]]
df_paid
```

```python
avg_subs_paid = round(df_paid["num_subscribers"].mean())
avg_subs_paid
```

```python
avg_subs_notpaid = round(df_notpaid["num_subscribers"].mean())
avg_subs_notpaid
```

```python
df['best-seller'] = df.apply(lambda row: 0, axis = 1)
for idx, row in df.iterrows():
    if  df.loc[idx,'is_paid'] == 1 and df.loc[idx,'num_subscribers'] > avg_subs_paid:
        df.loc[idx,'best-seller'] = 1
    if  df.loc[idx,'is_paid'] == 0 and df.loc[idx,'num_subscribers'] > avg_subs_notpaid:
        df.loc[idx,'best-seller'] = 1
df
```
